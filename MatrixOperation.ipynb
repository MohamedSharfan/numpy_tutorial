```python
import numpy as np
```

```python
q1 = np.array([
    [200,220,250],
    [150,180,210],
    [300,330,360]
])
q2 = np.array([
    [209,231,259],
    [155,192,222],
    [310,340,375]
])
```

add two

```python
q1 + q2
```

```python
q2 -q1
```

to find the growth percentage from q1 to q2

```python
 (q2 - q1)/q1 * 100
```

to find the revenue of the q1

```python
prices = np.array([
    [10,12,11], # prices for product a in different region
    [8,9,10],   #product b
    [15,16,17]  #product c
])
q1_revenue = q1 *prices
q1_revenue
```

if discount amount added

```python
q1_discounts = q1_revenue *20/100
q1_discounts
```

find net revenue (total revenue - discount)

```python
net_revenue = q1_revenue - q1_discounts
net_revenue
```

```python
np.sum(q1_discounts)
```

```python
np.sum(net_revenue)
```

dot product between 2 metrics


```python
features = np.array([
    [2000, 3], # house 1
    [1800, 4] # house 2
])

weights = np.array([150, 4000])
np.dot(features, weights) # house 1 = 2000*150 + 4000*3
```

cross product between 2 metrics

```python
c= np.array([1,2,3])
d = np.array([4,5,6])

np.cross(c,d)
```
